```python
import ast     
# Function to extract values from the Overview column
def extract_overview_values(overview_str):
    overview_list = ast.literal_eval(overview_str)
    return {item['key']: item['value'] for item in overview_list}

 # Apply the function to extract values and create new columns
overview_data = df['Overview'].apply(extract_overview_values).apply(pd.Series)

# Rename the new columns to avoid duplicates
overview_data.columns = [f'Overview_{col}' for col in overview_data.columns]

 # Concatenate the new columns to the original DataFrame
df = pd.concat([df, overview_data], axis=1)
      

```

```python

df=pd.read_csv('chennai4.csv')
df_head = df.head()
import pandas as pd
import ast
# Function to extract keys and values from Specifications_data
def extract_specifications(spec_data):
    try:
        specs = ast.literal_eval(spec_data)
        return [(item['key'], item['value']) for sublist in specs for item in sublist['list']]
    except:
        return []

# Create a new DataFrame to hold the extracted specifications
specifications = []
for index, row in df.iterrows():
    extracted = extract_specifications(row['Specifications_data'])
    specifications.append({'Model': row['Model'], **dict(extracted)})

specifications_df = pd.DataFrame(specifications)
specifications_head = specifications_df.head()
```
